```python
import pandas as pd
import os
```

```python
file = "U25 L3 Computer Science 電腦科學"
```

```python
text = '''
hacker
駭客

coding mistake
編碼錯誤

a bug in the code
代碼中的錯誤

work automatically without being encoded commands
自動工作無需編碼指令

mimic human cognition
模仿人類的認知

interpret external data
解釋外部數據

learning and problem solving
學習與解決問題

correctly interpret external data
正確解釋外部數據

learn to achieve specific goals
學習實現特定目標

definition changes
定義變更
'''
```

```python
file += ".csv"
file_path = f"data/{file}"

if file in os.listdir("data"):
    df0 = pd.read_csv(file_path)
else:
    df0 = pd.DataFrame(columns = ["English", "Chinese"])
```

```python
df0
```

```python
text = text.strip("\n")
words = [t.split("\n") for t in text.split("\n\n")]
df1 = pd.DataFrame(words, columns = ["English", "Chinese"])
df1
```

```python
df = pd.concat([df0, df1], ignore_index = True)
df
```

```python
df.to_csv(file_path, index = False)
```
